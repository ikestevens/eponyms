```python
from bs4 import BeautifulSoup
import requests
from ast import literal_eval
import pandas as pd
import urllib.request
import os
```

```python
root_path = 'https://art-ai.com/api/tokenInfo/'
image_root = 'https://ipfs.io/ipfs/'
```

```python
eponyms = {
    'token_num':[],
    "name":[],
    "image_url":[]
}
```

```python
for i in range(1, 10001):
    source = requests.get(root_path + str(i)).text
    soup = BeautifulSoup(source, 'lxml')
    ep_dict = literal_eval(soup.text)
    eponyms['token_num'].append(i)
    eponyms['name'].append(ep_dict['name'])
    eponyms['image_url'].append(image_root + ep_dict['image'].split("/")[2])
```

```python
#dataframe that has the | token number | name | image_url |
df = pd.DataFrame(eponyms)
```

```python
c = '\/:*?,<>|' #file names can't have these characters
```

```python
def downloadEponyms(row, path):
    if '\\' or '/' or ':' or "*" or "?"  or "," or "<" or ">" or "|" in row['name']:
        for char in c:
            row['name'] = row['name'].replace(char, '_')
    filename = str(row['name']) + '_' + str(row['token_num'])
    #download
    urllib.request.urlretrieve(row['image_url'], path + filename + '.jpg')
```

```python
#download every eponym
full_path = os.getcwd() + '/images/'
count = 1
for index, row in df.iterrows():
    if count < 2408:
        count+=1
        continue
    if count == 2408: #2408 just doesn't work, there's no pic there
        count+=1
        continue
    downloadEponyms(row, full_path)
    print('Downloaded {} eponyms'.format(count))
    count+=1
```
